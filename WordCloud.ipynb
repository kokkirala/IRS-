```python
from wordcloud import WordCloud
text_data = "Imagine you're a marketing manager for a leading tech company. During a recent product launch, you collected feedback from customers through online surveys, social media channels, and email correspondence. After compiling all the data, you decide to create a word cloud to visually represent the most common sentiments and feedback from your customers."
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_data)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
```
